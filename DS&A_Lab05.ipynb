```python
X = data["YearsExperience"]
y = data["Salary"]
```

```python
X.head()
```

```python
y.head(1)
```

```python
plt.scatter(X,y)
plt.title("Salary - Experience in Years")
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.show()
```

```python
model = LinearRegression()
X = X.values.reshape(-1,1)
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict(np.array([[5], [10], [20]]))
```

```python
x1 = sm.add_constant(X)
results = sm.OLS(y,x1).fit()
results.summary()
```
